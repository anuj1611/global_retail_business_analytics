**Cell A — Prep dataframe for SQL (normalize column names, parse dates, add derived columns)**

This makes column names SQL-friendly (order_date instead of Order Date) and creates shipping_lead_days (ship_date - order_date) which is useful for delivery/performance analysis.

```python
# A - Normalize columns, parse dates, create derived columns
import pandas as pd
import numpy as np
import os

# load the UTF-8 saved file (from previous step)
utf8_path = '/content/data/Global_Superstore_UTF8.csv'
df = pd.read_csv(utf8_path, encoding='utf-8')

# 1. Normalize column names to snake_case (SQL-friendly)
def normalize_col(c):
    c = c.strip()
    c = c.replace(' ', '_').replace('-', '_').replace('/', '_')
    c = ''.join(ch for ch in c if ch.isalnum() or ch == '_')  # remove odd chars
    return c.lower()

df.columns = [normalize_col(c) for c in df.columns]

# 2. Parse date columns (common names in this dataset)
date_cols = []
for c in ['order_date', 'ship_date', 'delivery_date', 'order_date_time']:
    if c in df.columns:
        date_cols.append(c)

# Try to parse any found date columns
for c in date_cols:
    df[c] = pd.to_datetime(df[c], errors='coerce', dayfirst=False)

# 3. Create derived columns
if 'order_date' in df.columns and 'ship_date' in df.columns:
    df['shipping_lead_days'] = (df['ship_date'] - df['order_date']).dt.days
else:
    df['shipping_lead_days'] = np.nan

# 4. Trim whitespace in string columns and fix simple types
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.strip()

# 5. Convert numeric columns safely
numeric_candidates = ['sales','profit','shipping_cost','quantity','discount']
for c in numeric_candidates:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

# Quick sanity checks
print("Preview after preprocessing:")
print("Shape:", df.shape)
print("Some columns:", df.columns.tolist()[:15])
df.head()
```

**Cell D — Useful exploratory SQL queries (run any of these)**

It has queries and Python code that runs them and prints results. Copy any SQL snippet into the sql variable and run the cell.

```python
# D - Run example SQL queries and display results quickly

def run_sql(q, top=10):
    print("SQL QUERY:")
    print(q)
    print("\nRESULT:")
    df_q = pd.read_sql_query(q, conn)
    display(df_q.head(top))
    print("-" * 60)
    return df_q

# 1. Quick row count and date range
q1 = f"SELECT COUNT(*) AS total_rows, MIN(order_date) AS first_order, MAX(order_date) AS last_order FROM {table_name};"
run_sql(q1)

# 2. Basic KPIs: total sales, orders count, avg order value
# Note: order_id may not be unique per row; adjust if dataset uses order_item_id
q2 = f"""
SELECT
  COUNT(DISTINCT order_id) AS orders,
  SUM(sales) AS total_sales,
  AVG(sales) AS avg_row_sales -- average per row; not per-order
FROM {table_name};
"""
run_sql(q2)

# 3. Revenue by region (top 10)
q3 = f"""
SELECT region, COUNT(*) AS rows, SUM(sales) AS revenue, AVG(profit) AS avg_profit
FROM {table_name}
GROUP BY region
ORDER BY revenue DESC
LIMIT 10;
"""
run_sql(q3)

# 4. Top 10 products by revenue
q4 = f"""
SELECT product_name, product_id, SUM(sales) AS revenue, SUM(quantity) AS qty_sold
FROM {table_name}
GROUP BY product_id, product_name
ORDER BY revenue DESC
LIMIT 10;
"""
run_sql(q4)

# 5. Monthly sales trend (by order_date)
# Make sure order_date exists and is stored in ISO format in sqlite
if 'order_date' in df.columns:
    q5 = f"""
    SELECT strftime('%Y-%m', order_date) AS order_month,
           COUNT(DISTINCT order_id) AS orders,
           SUM(sales) AS total_sales,
           AVG(shipping_lead_days) AS avg_lead_days
    FROM {table_name}
    GROUP BY order_month
    ORDER BY order_month;
    """
    run_sql(q5)

# 6. Shipping lead days distribution and slow carriers/regions
q6 = f"""
SELECT region,
       COUNT(*) AS orders_count,
       AVG(shipping_lead_days) AS avg_lead_days,
       SUM(CASE WHEN shipping_lead_days > 7 THEN 1 ELSE 0 END) AS orders_over_7_days
FROM {table_name}
GROUP BY region
ORDER BY avg_lead_days DESC
LIMIT 20;
"""
run_sql(q6)

# 7. Orders causing loss (negative profit)
q7 = f"""
SELECT COUNT(*) AS loss_orders, SUM(profit) AS total_loss
FROM {table_name}
WHERE profit < 0;
"""
run_sql(q7)

# 8. High-level returns/returns proxy (if 'order_status' or 'is_return' exists)
if 'order_status' in df.columns:
    q8 = f"""
    SELECT order_status, COUNT(*) AS cnt
    FROM {table_name}
    GROUP BY order_status
    ORDER BY cnt DESC;
    """
    run_sql(q8)
else:
    print("No order_status column found; skip returns summary.")
```

**Cell 4A — Quick data-quality summary (baseline)**

Purpose: understand where problems exist (missing values, types, unique counts, ranges) before changing anything.

This produces a table listing each column’s dtype, null counts, unique counts and some samples so we know where to focus cleaning.

```python
# 4A - Data quality baseline
import pandas as pd, numpy as np
utf8_path = '/content/data/Global_Superstore_UTF8.csv'
df = pd.read_csv(utf8_path, encoding='utf-8')

def dq_summary(df):
    cols = df.columns.tolist()
    d = []
    for c in cols:
        col = df[c]
        dtype = col.dtype
        nnull = col.isna().sum()
        nunique = col.nunique(dropna=True)
        sample_vals = col.dropna().astype(str).head(3).tolist()
        d.append({
            'column': c,
            'dtype': str(dtype),
            'non_null%': round(100*(len(col)-nnull)/len(col),2),
            'n_null': int(nnull),
            'n_unique': int(nunique),
            'sample_vals': sample_vals
        })
    return pd.DataFrame(d).sort_values('non_null%', ascending=True)

dq = dq_summary(df)
print("Rows:", len(df))
display(dq)
# quick numeric overview for candidate numeric columns
num_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c]) or c.lower() in ['sales','profit','quantity','discount','shipping_cost']]
if num_cols:
    display(df[num_cols].describe().T)
```

**Cell 4B — Normalize column names again & basic trimming**

Purpose: ensure column names are stable and there are no leading/trailing spaces in string data.

Explanation:

Removes stray spaces in column names and values; treats empty strings as missing.

```python
# 4B - Normalize column names and trim strings
def normalize_col(c):
    c = str(c).strip()
    c = c.replace(' ', '_').replace('-', '_').replace('/', '_')
    c = ''.join(ch for ch in c if ch.isalnum() or ch == '_')
    return c.lower()

df.columns = [normalize_col(c) for c in df.columns]

# Trim whitespace for object columns and replace empty strings with NaN
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str).str.strip().replace({'': pd.NA, 'nan': pd.NA})

print("Normalized columns and trimmed strings. Sample columns:", df.columns.tolist()[:10])
```

**Cell 4C — Parse date columns robustly and add derived date parts**

Purpose: ensure date columns are real datetimes and create year/month/day columns for grouping.

Explanation:

We search for columns that include 'date', parse them to datetime (coerce invalids to NaT), and add year/month for grouping. This prevents later surprises when grouping by month in SQL/Power BI.

```python
# 4C - Parse dates and add date parts
date_candidates = [c for c in df.columns if 'date' in c.lower()]
print("Date candidate columns:", date_candidates)

for c in date_candidates:
    df[c+'_parsed'] = pd.to_datetime(df[c], errors='coerce', dayfirst=False)  # coerce bad values to NaT

# Standardize main date columns (rename parsed if present)
# pick the most important date columns if they exist
if 'order_date_parsed' in df.columns:
    df['order_date'] = df['order_date_parsed']
if 'ship_date_parsed' in df.columns:
    df['ship_date'] = df['ship_date_parsed']
if 'delivery_date_parsed' in df.columns:
    df['delivery_date'] = df['delivery_date_parsed']

# Create year/month/day and month_start for time grouping
if 'order_date' in df.columns:
    df['order_year'] = df['order_date'].dt.year
    df['order_month'] = df['order_date'].dt.to_period('M').dt.to_timestamp()
    df['order_day'] = df['order_date'].dt.day

print("Parsed dates and created year/month/day columns. Example:")
display(df[['order_date','order_year','order_month']].head(3))
```

**Cell 4D — Numeric coercion and obvious numeric fixes**

Purpose: ensure sales/profit/quantity etc. are numeric and handle obvious nonsense values.

Explanation:

Converts numeric-looking cols to numbers. Flags negative quantity/sales (may indicate returns or bad records) for later handling.

```python
# 4D - Numeric coercion for common numeric columns
numeric_candidates = ['sales','profit','quantity','discount','shipping_cost']
for c in numeric_candidates:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

# Basic checks & fixes:
# - Negative quantities -> treat as suspect (flag)
# - Zero or negative prices: flag for inspection (some returns/refunds can be negative profit)
df['negative_quantity_flag'] = False
if 'quantity' in df.columns:
    df['negative_quantity_flag'] = df['quantity'] < 0

df['negative_sales_flag'] = False
if 'sales' in df.columns:
    df['negative_sales_flag'] = df['sales'] < 0

display(df[numeric_candidates + ['negative_quantity_flag','negative_sales_flag']].describe().T)
```

**Cell 4F — Handle missing values (strategy per column type)**

Purpose: decide how to treat missing values: drop, fill, or flag.

Explanation:

We report missing counts. For truly critical columns we drop rows (order_id, order_date, sales). For categorical descriptions we fill with 'Unknown' to preserve rows.

```python
# 4F - Missing value strategy
miss_report = df.isna().sum().sort_values(ascending=False)
display(miss_report[miss_report > 0].to_frame('n_missing').head(50))

# Strategy (examples)
# - Critical columns (order_id, order_date, sales): drop rows if missing
critical_cols = [c for c in ['order_id','order_date','sales'] if c in df.columns]
print("Critical cols:", critical_cols)
before = len(df)
df = df.dropna(subset=critical_cols)
print(f"Dropped rows missing critical columns: {before - len(df)} rows removed.")

# - Low-impact columns: fill with sentinel or 'Unknown'
for c in ['region','country','segment','product_name']:
    if c in df.columns:
        df[c] = df[c].fillna('Unknown')
```

**Cell 4G — Handle outliers for numeric columns (IQR flagging, not automatic removal)**

Purpose: detect extreme outliers so you can decide case-by-case (flagging is safer than automatic deletion).

Explanation:

Flags rows where numeric values are outside typical range by IQR method. You can review flagged rows manually; automatic trimming can distort analysis.

```python
# 4G - Outlier detection using IQR (flagging)
def flag_outliers_iqr(s, k=1.5):
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    low = q1 - k * iqr
    high = q3 + k * iqr
    return (s < low) | (s > high)

for c in ['sales','profit','shipping_cost','quantity']:
    if c in df.columns:
        flag_col = c + '_outlier'
        df[flag_col] = flag_outliers_iqr(df[c].dropna())
        print(f"Outlier flag created for {c}: sample count = {df[flag_col].sum() if flag_col in df.columns else 'N/A'}")
```

**Cell 4H — Create business flags (is_return, is_discounted, on_time)**

Purpose: make common business logic columns that will be used in KPIs.


Explanation:

Adds practical boolean flags used in dashboards: returns, discounted orders, on-time deliveries. You may refine SLA rules later.

```python
# 4H - Business flags
# is_return: try to detect using order_status or negative_sales_flag or product_name containing 'Return'
df['is_return'] = False
if 'order_status' in df.columns:
    df['is_return'] = df['order_status'].str.contains('Return', case=False, na=False)
else:
    # fallback: negative sales or negative quantity may indicate return/refund
    if 'negative_sales_flag' in df.columns:
        df['is_return'] = df['negative_sales_flag']

# is_discounted
df['is_discounted'] = False
if 'discount' in df.columns:
    df['is_discounted'] = df['discount'].fillna(0) > 0

# on_time (if delivery_date exists)
df['on_time'] = pd.NA
if 'delivery_date' in df.columns and 'ship_date' in df.columns:
    # on_time = delivery_date <= ship_date + expected_sla (we don't have expected_sla, so use delivery <= ship)
    df['delivery_delay_days'] = (df['delivery_date'] - df['ship_date']).dt.days
    df['on_time'] = df['delivery_delay_days'].le(0)  # delivered on same day or earlier than ship_date
else:
    if 'shipping_lead_days' in df.columns:
        df['on_time'] = df['shipping_lead_days'].le(0)
```

**Cell 5C — Performance KPIs (Returns, Discounts, On-Time)**

Explanation:

Uses Boolean flags created in Step 4 (is_return, is_discounted, on_time).

Calculates percentages over total rows — gives high-level operational efficiency.



**Note:**  
The dataset does not contain a `Delivery Date` column, so actual delivery delays could not be computed.  
Instead, we derived a `shipping_lead_days` metric (`Ship Date - Order Date`) to analyze order fulfillment speed.  
A derived flag `fast_ship` (≤ 1 day) was used as a proxy for "on-time delivery".


```python
# Updated 5C: Performance KPIs (handles missing columns safely)
import pandas as pd

# Check if delivery_delay_days column exists
cols_df = pd.read_sql_query("PRAGMA table_info(orders_clean);", conn)
cols = cols_df['name'].tolist()

if 'delivery_delay_days' in cols:
    q_perf = """
    SELECT
      ROUND(SUM(CASE WHEN is_return=1 THEN 1 ELSE 0 END)*100.0/COUNT(*), 2) AS return_rate_pct,
      ROUND(SUM(CASE WHEN is_discounted=1 THEN 1 ELSE 0 END)*100.0/COUNT(*), 2) AS discount_rate_pct,
      ROUND(SUM(CASE WHEN on_time=1 THEN 1 ELSE 0 END)*100.0/COUNT(*), 2) AS on_time_delivery_pct,
      ROUND(AVG(delivery_delay_days), 2) AS avg_delivery_delay_days
    FROM orders_clean;
    """
else:
    q_perf = """
    SELECT
      ROUND(SUM(CASE WHEN is_return=1 THEN 1 ELSE 0 END)*100.0/COUNT(*), 2) AS return_rate_pct,
      ROUND(SUM(CASE WHEN is_discounted=1 THEN 1 ELSE 0 END)*100.0/COUNT(*), 2) AS discount_rate_pct,
      ROUND(SUM(CASE WHEN on_time=1 THEN 1 ELSE 0 END)*100.0/COUNT(*), 2) AS on_time_delivery_pct
    FROM orders_clean;
    """

perf_kpis = pd.read_sql_query(q_perf, conn)
print("Performance KPIs:")
display(perf_kpis)
```

```python
# Regional Sales:
q_region = """
SELECT
  region,
  ROUND(SUM(sales),2) AS total_sales,
  ROUND(SUM(profit),2) AS total_profit,
  ROUND(AVG(profit/sales)*100,2) AS avg_margin_pct,
  ROUND(SUM(CASE WHEN on_time=1 THEN 1 ELSE 0 END)*100.0/COUNT(*),2) AS on_time_pct
FROM orders_clean
GROUP BY region
ORDER BY total_sales DESC;
"""
region_summary = pd.read_sql_query(q_region, conn)
print("Regional Summary:")
display(region_summary)


# Top 10 Products by Revenue:
q_products = """
SELECT
  product_name,
  ROUND(SUM(sales),2) AS total_sales,
  ROUND(SUM(profit),2) AS total_profit,
  COUNT(DISTINCT order_id) AS order_count
FROM orders_clean
GROUP BY product_name
ORDER BY total_sales DESC
LIMIT 10;
"""
top_products = pd.read_sql_query(q_products, conn)
print("Top 10 Products by Revenue:")
display(top_products)
```

```python
# --- STEP 6: Fix flags based on actual CSV structure ---
import sqlite3, pandas as pd
conn = sqlite3.connect('/content/data/fpadb.sqlite')

df = pd.read_sql_query("SELECT * FROM orders_clean;", conn)

# is_return: detect rows with negative sales or profit
df['is_return'] = 0
if 'sales' in df.columns:
    df['is_return'] = (pd.to_numeric(df['sales'], errors='coerce') < 0).astype(int)
elif 'profit' in df.columns:
    df['is_return'] = (pd.to_numeric(df['profit'], errors='coerce') < 0).astype(int)

# is_discounted: discount > 0
df['is_discounted'] = 0
if 'discount' in df.columns:
    df['is_discounted'] = (pd.to_numeric(df['discount'], errors='coerce') > 0).astype(int)

# shipping_lead_days = ship_date - order_date
if 'shipping_lead_days' not in df.columns:
    if 'order_date' in df.columns and 'ship_date' in df.columns:
        df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
        df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce')
        df['shipping_lead_days'] = (df['ship_date'] - df['order_date']).dt.days
    else:
        df['shipping_lead_days'] = pd.NA

# fast_ship proxy: shipped within 1 day
df['fast_ship'] = (pd.to_numeric(df['shipping_lead_days'], errors='coerce') <= 1).fillna(0).astype(int)

df.to_sql('orders_clean', conn, if_exists='replace', index=False)
print(" Updated orders_clean with is_return, is_discounted, shipping_lead_days, fast_ship")
```

```python
# --- STEP 6: Region-level Return vs Fast Ship chart ---
import matplotlib.pyplot as plt

q = """
SELECT region,
       SUM(CASE WHEN is_return=1 THEN 1 ELSE 0 END) AS n_returns,
       SUM(CASE WHEN fast_ship=1 THEN 1 ELSE 0 END) AS n_fast_ship,
       COUNT(*) AS n_rows
FROM orders_clean
GROUP BY region
ORDER BY n_rows DESC;
"""
df_reg = pd.read_sql_query(q, conn)
df_reg['return_rate_pct'] = 100.0 * df_reg['n_returns'] / df_reg['n_rows']
df_reg['fast_ship_pct']   = 100.0 * df_reg['n_fast_ship'] / df_reg['n_rows']

display(df_reg[['region','n_rows','return_rate_pct','fast_ship_pct']])

# Plot
x = range(len(df_reg))
width = 0.35
fig, ax = plt.subplots(figsize=(12,5))
ax.bar([p - width/2 for p in x], df_reg['return_rate_pct'], width=width, label='Return Rate (%)')
ax.bar([p + width/2 for p in x], df_reg['fast_ship_pct'], width=width, label='Fast Ship (%)')
ax.set_xticks(x)
ax.set_xticklabels(df_reg['region'], rotation=45, ha='right')
ax.set_ylabel('Percentage (%)')
ax.set_title('Return Rate vs Fast Ship (%) by Region (Fast Ship = shipped ≤ 1 day after order)')
ax.legend()
plt.tight_layout()
plt.show()
```
